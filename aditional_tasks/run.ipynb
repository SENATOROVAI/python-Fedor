#### Создать граф и свизуализировать на матплотлиб , networkx

```python
import networkx as nx

G = nx.Graph()  # объект класса networkx.classes.graph.Graph
print(G)

# добавим узлы без ребер
# .add_node добавляет узлы
G.add_node("Singapore")
G.add_node("San Francisco")
G.add_node("Tokyo")

# с помощью .add_nodes_from([item1,item2,item3]) можно добавть сразу множество узлов
G.add_nodes_from(["Riga", "Copenhagen"])
print(G)

# когда узлы определены, нужно определить ребра и соеденить их передав узлы в функцию .add_edge()
# или передать кортеж в функцию .add_edges_from()
G.add_edge("Singapore", "San Francisco")
G.add_edge("San Francisco", "Tokyo")
G.add_edges_from(
    [
        ("Riga", "Copenhagen"),
        ("Copenhagen", "Singapore"),
        ("Singapore", "Tokyo"),
        ("Riga", "San Francisco"),
        ("San Francisco", "Singapore"),
    ]
)

# Отображение меток
# отрисовка осуществляется с помощью nx.draw()
nx.draw(G)
```
