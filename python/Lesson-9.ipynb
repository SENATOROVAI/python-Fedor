### Decorators


```python
def foo(x):
    return x

def power(f):
    def wrapper(x):
        result=f(x*x)
        return result
    return wrapper

d = power(foo)
print(d(5))
```

```python
def hello():
    return 'hello word'

def uppercase(func):
    def wrapper():
        res = func().upper()
        return res
    return wrapper

def add_stars(func):
    def wrapper():
        res = func()+'*'*5
        return res
    return wrapper

d = add_stars(uppercase(hello))
print(d())

```

```python
@add_stars
@uppercase
def hello():
    return 'hello world'

print(hello())
```

```python
def add_number(func):
    def wrapper(*args,**kwargs):
        print(f'Call_Function: {func.__name__}')
        print(f'Arguments: {args,kwargs}')
        result = func(*args,**kwargs)
        print(f'Result: {result}')
    return wrapper

@add_number
def summa(a,b):
    return a+b

# d = add_number(summa(1,4))
# print(d(1,23,4,1,2,3))
```

```python
def calls_count(func):
    count = 0
    def wrapper(*args,**kwargs):
        nonlocal count
        count+=1
        print(f'{count}')
        return func(*args,**kwargs)
    return wrapper

@calls_count
def power():
    return 'HelloWorld'

```
