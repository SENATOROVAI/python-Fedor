Задание 3 (0.5)
Напишите программу, которая конвертирует строковые значения данного словаря nums в числа, там, где это применимо.

```python
d = { 'x':'10' , 'y':'20' , 'z':'30', 'p':'40', 'q':'50', 'r':'60'}
n = {}
for k, v in d.items():
    try:
        v = int(v)
    except :
        pass
    n[k] = v
print(n)



# { 'x': 10 , 'y': 20 , 'z': 30, 'p': 40, 'q': 50, 'r': 60}
```
